```python
import matplotlib.pyplot as plt
import numpy as np
import h5py
```

```python
with h5py.File('./data/train_data.hdf5', 'r') as h5_file:
    train_data = h5_file['true_energies'][()]
fig, ax = plt.subplots(1, 2, figsize=(15, 8))
ax[0].scatter(np.arange(len(train_data)), train_data)
ax[1].hist(train_data - np.mean(train_data))
```
